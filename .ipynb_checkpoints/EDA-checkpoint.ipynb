# Análise Exploratória de Dados

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import scipy.stats as sct
```

```python
df = pd.read_csv('casas_ponta_grossa.csv')
```

```python
df.info()
```

```python
df.describe()
```

## Relações entre as features discretas e o Valor de Venda

```python
plt.figure(figsize = (8,6))
plt.title('Quartos e Valor de Venda')
sns.barplot(x = 'Quartos', y = 'Valor Venda', data = df)
plt.show()
```

```python
plt.figure(figsize = (8,6))
plt.title('Banheiros e Valor de Venda')
sns.barplot(x = 'Banheiros', y = 'Valor Venda', data = df)
plt.show()
```

```python
plt.figure(figsize = (8,6))
plt.title('Vagas de Garagem e Valor de Venda')
sns.barplot(x = 'Vagas de Garagem', y = 'Valor Venda', data = df)
plt.show()
```

# Distribuição das variáveis contínuas

## Área Total:

```python
print("Skewness: ", df['Área Total'].skew())
print("Curtose: ", df['Área Total'].kurt()) # Curtose é muito sensível a outliers
```

```python
plt.figure(figsize = (8,6))
plt.title('Distribuição da Área Total')
sns.distplot(df['Área Total']);
```

### Área Total, análise retirando alguns outliers:

```python
print("Skewness: ", df[df['Área Total']<1000]['Área Total'].skew())
print("Curtose: ", df[df['Área Total']<1000]['Área Total'].kurt()) 
```

```python
plt.figure(figsize = (8,6))
plt.title('Distribuição da Área Total Retirando Outliers')
sns.distplot(df[df['Área Total']<1000]['Área Total']);
```

## Valor de Venda:

```python
print("Skewness: ", df['Valor Venda'].skew())
print("Curtose: ", df['Valor Venda'].kurt())
```

```python
plt.figure(figsize = (8,6))
sns.distplot(df['Valor Venda']);
```

# Analisando o valor médio das casas pro bairro

```python
plt.figure(figsize = (8,12))
plt.title('Número de casas por bairro')
sns.countplot(y = 'Bairro', data = df)
plt.show()
```

```python
plt.figure(figsize = (15,15))
plt.title('Valor médio das casas por bairro')
sns.barplot(x = 'Valor Venda', y = 'Bairro', data = df)
plt.show()
```

# Correlação entre as features

```python
corr = df.corr()
figure, ax = plt.subplots(figsize=(8,8))
plt.title('Correlação entre as variáveis')
sns.heatmap(corr, annot=True, square=True);
```

# Scatter plots entre features e valor de venda 

```python
plt.figure(figsize = (12,8))
plt.title('Scatter plot entre Valor de Venda e nº de banheiros')
banheiro_valor_venda = pd.concat([df['Quartos'],df['Valor Venda']],axis = 1)
sns.regplot(x='Quartos',y = 'Valor Venda',data = banheiro_valor_venda, scatter= True, fit_reg=True)
```

```python
plt.figure(figsize = (12,8))
plt.title('Scatter plot entre Valor de Venda e nº de banheiros')
banheiro_valor_venda = pd.concat([df['Banheiros'],df['Valor Venda']],axis = 1)
sns.regplot(x='Banheiros',y = 'Valor Venda',data = banheiro_valor_venda, scatter= True, fit_reg=True, color='g')
```

```python
plt.figure(figsize = (12,8))
plt.title('Scatter plot entre Valor de Venda e as vagas de garagem')
banheiro_valor_venda = pd.concat([df['Vagas de Garagem'],df['Valor Venda']],axis = 1)
sns.regplot(x='Vagas de Garagem',y = 'Valor Venda',data = banheiro_valor_venda, scatter= True, fit_reg=True, color='y')
```

```python
plt.figure(figsize = (12,8))
plt.title('Scatter plot entre Valor de Venda e o nº de Suítes')
banheiro_valor_venda = pd.concat([df['Suítes'],df['Valor Venda']],axis = 1)
sns.regplot(x='Suítes',y = 'Valor Venda',data = banheiro_valor_venda, scatter= True, fit_reg=True, color="purple")
```
